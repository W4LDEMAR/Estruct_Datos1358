```python
# Estructura de datos
# Sosa Hernandez Saul Waldemar

class BinaryNode:
  def __init__(self, value, left = None, right = None):
    self.data = value
    self.left = left
    self.right = right
  
class BinarySearchTree:
  def __init__(self):
    self.root = None

  def insert(self, value):
    if self.root == None:
      self.root = BinaryNode(value)
    else:
      self.__insert_nodo(self.root, value)

  def __insert_nodo(self, nodo, value):
    if nodo.data == value:
      pass
    elif value < nodo.data:
      if nodo.left == None:
        nodo.left = BinaryNode(value)
      else:
        self.__insert_nodo(nodo.left, value)
    else:
      if nodo.right == None:
        nodo.right = BinaryNode(value)
      else:
        self.__insert_nodo(nodo.right, value)
  
  def transversal(self, formato = "posorden"):
    if formato == "posorden":
      self.recorrido_pos(self.root)
    elif formato == "preorden":
      self.recorrido_pre(self.root)
    else:
      self.recorrido_in(self.root)
    print("")

  def recorrido_pos(self, nodo):
    if nodo != None:
      self.recorrido_pos(nodo.left)
      self.recorrido_pos(nodo.right)
      print(str(nodo.data), end = " , ")

  def recorrido_pre(self, nodo):
    if nodo != None:
      print(str(nodo.data), end = " , ")
      self.recorrido_pre(nodo.left)
      self.recorrido_pre(nodo.right)

  def recorrido_in(self, nodo):
    if nodo != None:
      self.recorrido_in(nodo.left)
      print(str(nodo.data), end = " , ")
      self.recorrido_in(nodo.right)

  def search(self, value):
    if self.root == None:
      print("El arbol esta vacio")
      return None
    else:
      return self.__search(self.root, value)

  def __search(self, nodo, value):
    if nodo == None:
      print("El dato no existe en el arbol")
      return None
    elif nodo.data == value:
      print("Encontrado")
      return nodo
    elif value < nodo.data:
      return self.__search(nodo.left, value)
    else:
      return self.__search(nodo.right, value)

  def remove(self, value):
    if self.root == None:
      print("El arbol esta vacio")
      return None
    else:
      return self.__remove(self.root, value)
  
  def reemplazo(self, nodo):
    hoja = nodo.data
    while nodo.left != None:
      hoja = nodo.left.data
      nodo = nodo.left
    return hoja
  
  def __remove(self, nodo, value):
    if nodo == None:
      print("El dato no existe en el arbol.")
      return None
    elif value < nodo.data:
      nodo.left = self.__remove(nodo.left, value)
    elif value > nodo.data:
      nodo.right = self.__remove(nodo.right, value)
    else:
      if nodo.left == None:
        return nodo.right
      elif nodo.right == None:
        return nodo.left
      nodo.data = self.reemplazo(nodo.right)
      nodo.right = self.__remove(nodo.right, nodo.data)
    return nodo
```

```python
ejemplo = BinarySearchTree()
ejemplo.insert(8)
ejemplo.insert(5)
ejemplo.insert(1)
ejemplo.insert(4)
ejemplo.insert(6)
ejemplo.insert(7)
ejemplo.insert(20)
ejemplo.insert(12)
ejemplo.insert(23)
ejemplo.transversal("preorden")
print("---------------------------------------")
ejemplo.remove(45) #Un dato que no existe
ejemplo.remove(7)
ejemplo.remove(23)
ejemplo.remove(4)
ejemplo.transversal("preorden")
```
